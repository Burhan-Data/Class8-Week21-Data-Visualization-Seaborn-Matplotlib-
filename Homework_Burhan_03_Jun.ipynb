#### Features

- timestamp - timestamp field for grouping the data
- cnt - the count of a new bike shares
- t1 - real temperature in C
- t2 - temperature in C “feels like”
- hum - humidity in percentage
- wind_speed - wind speed in km/h
- weather_code - category of the weather
- is_holiday - boolean field - 1 holiday / 0 non holiday
- is_weekend - boolean field - 1 if the day is weekend
- season - category field meteorological seasons: 0-spring ; 1-summer; 2-fall; 3-winter.

**"weather_code" category description:**
* 1 = Clear ; mostly clear but have some values with haze/fog/patches of fog/ fog in vicinity 
* 2 = scattered clouds / few clouds 
* 3 = Broken clouds 
* 4 = Cloudy 
* 7 = Rain/ light Rain shower/ Light rain 
* 10 = rain with thunderstorm 
* 26 = snowfall 
* 94 = Freezing Fog

### Visualize the correlation with a heatmap

```python
# for all variables 
corr_matrix = df.corr()
```

```python
plt.figure(figsize=(10, 8))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', linewidths=0.5)
plt.title('Correlation Heatmap')
plt.show()
```

```python
# for the target variable
corr_matrix = df[['cnt', 't1', 't2', 'hum']].corr()
plt.figure(figsize=(10, 8))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', linewidths=0.5)
plt.title('Correlation Heatmap')
plt.show()
```

### Visualize the correlation of the target variable and the other features with barplot

```python
#with pandas
corr_matrix = df[['cnt', 't1', 't2', 'hum']].corr()
corr_matrix.plot(kind='bar', figsize=(10, 6), color='steelblue')

# Set the plot title and axis labels
plt.title('Correlation with Target Variable')
plt.xlabel('Features')
plt.ylabel('Correlation')
```

### Plot bike shares by months and year_of_month (use lineplot, pointplot, barplot)

```python
df_monthly = df['cnt'].resample('M').mean()
df_year_monthly = df['cnt'].resample('M').mean()
```

```python
# with lineplot

plt.figure(figsize=(10, 6))
sns.lineplot(data=df_monthly)
plt.title('Lineplot of cnt by Month')
plt.xlabel('Month')
plt.ylabel('cnt')
plt.show()
```

```python
# with pointplot
plt.figure(figsize=(10, 6))
sns.pointplot(data=df_monthly)
plt.title('Pointplot of cnt by Month')
plt.xlabel('Month')
plt.ylabel('cnt')
plt.show()
```

```python
# with barplot
plt.figure(figsize=(10, 6))
sns.barplot(x=df_year_monthly.index.strftime('%Y-%m'), y=df_year_monthly)
plt.title('Barplot of cnt by Year and Month')
plt.xlabel('Year and Month')
plt.ylabel('cnt')
plt.xticks(rotation=45)
plt.show()
```

### Plot bike shares by hours on (holidays, weekend, season)

```python
# with lineplot (*whether it is a holiday or not*)
df_hourly = df[['cnt', 'is_holiday', 'is_weekend', 'season']].resample('H').mean()

```

```python
df_hourly.head()
```

```python
sns.set(style="darkgrid")
fig, axes = plt.subplots(3, 1, figsize=(10, 15), sharex=True)
# Plot bike shares by hours on holidays
sns.lineplot(data=df_hourly[df_hourly['is_holiday'] == 1], x='timestamp', y='cnt', hue='is_holiday', ax=axes[0])
axes[0].set_title('Bike Shares by Hours on Holidays')
```
